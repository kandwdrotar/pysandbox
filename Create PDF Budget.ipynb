```python
#Income - Net

##Single federal tax marginal rate

s_net_salary = 0

if((filing_status == 'S' or filing_status == 's') and self_income <= s_max1):
    s_net_salary = ((self_income)*(1-fica))-(s_1*(self_income))
elif((filing_status == 'S' or filing_status == 's') and self_income <= s_max2):
    s_net_salary = ((self_income)*(1-fica))-(s_1*s_max1)-(((self_income)-s_max1)*s_2)
elif((filing_status == 'S' or filing_status == 's') and self_income <= s_max3):
    s_net_salary = ((self_income)*(1-fica))-(s_1*s_max1)-(s_2*(s_max2-s_max1))-(((self_income)-s_max2)*s_3)
elif((filing_status == 'S' or filing_status == 's') and self_income <= s_max4):
    s_net_salary = ((self_income)*(1-fica))-(s_1*s_max1)-(s_2*(s_max2-s_max1))-(s_3*(s_max3-s_max2))-(((self_income)-s_max3)*s_4)

    
s_net_monthly = s_net_salary/12
s_net_weekly = s_net_salary/52
s_net_hourly = s_net_salary/52/40
s_net_salary = s_net_salary

##Married Filing Jointly federal tax marginal rate

mfj_net_salary = 0

if((filing_status == 'MFJ' or filing_status == 'mfj') and (self_income + spouse_income) <= mfj_max1):
    mfj_net_salary = ((self_income + spouse_income)*(1-fica))-(mfj_1*(self_income + spouse_income))
elif((filing_status == 'MFJ' or filing_status == 'mfj') and (self_income + spouse_income) <= mfj_max2):
    mfj_net_salary = ((self_income + spouse_income)*(1-fica))-(mfj_1*mfj_max1)-(((self_income + spouse_income)-mfj_max1)*mfj_2)
elif((filing_status == 'MFJ' or filing_status == 'mfj') and (self_income + spouse_income) <= mfj_max3):
    mfj_net_salary = ((self_income + spouse_income)*(1-fica))-(mfj_1*mfj_max1)-(mfj_2*(mfj_max2-mfj_max1))-(((self_income + spouse_income)-mfj_max2)*mfj_3)
elif((filing_status == 'MFJ' or filing_status == 'mfj') and (self_income + spouse_income) <= mfj_max4):
    mfj_net_salary = ((self_income + spouse_income)*(1-fica))-(mfj_1*mfj_max1)-(mfj_2*(mfj_max2-mfj_max1))-(mfj_3*(mfj_max3-mfj_max2))-(((self_income + spouse_income)-mfj_max3)*mfj_4)

    
mfj_net_monthly = mfj_net_salary/12
mfj_net_weekly = mfj_net_salary/52
mfj_net_hourly = mfj_net_salary/52/40
mfj_net_salary = mfj_net_salary

##Married Filing Separately federal tax marginal rate    

mfs_net_salary = 0

if((filing_status == 'MFS' or filing_status == 'mfs') and self_income <= s_max1):
    s_net_salary = ((self_income)*(1-fica))-(s_1*(self_income))
elif((filing_status == 'MFS' or filing_status == 'mfs') and self_income <= s_max2):
    s_net_salary = ((self_income)*(1-fica))-(s_1*s_max1)-(((self_income)-s_max1)*s_2)
elif((filing_status == 'MFS' or filing_status == 'mfs') and self_income <= s_max3):
    s_net_salary = ((self_income)*(1-fica))-(s_1*s_max1)-(s_2*(s_max2-s_max1))-(((self_income)-s_max2)*s_3)
elif((filing_status == 'MFS' or filing_status == 'mfs') and self_income <= s_max4):
    s_net_salary = ((self_income)*(1-fica))-(s_1*s_max1)-(s_2*(s_max2-s_max1))-(s_3*(s_max3-s_max2))-(((self_income)-s_max3)*s_4)

    
mfs_net_monthly = mfs_net_salary/12
mfs_net_weekly = mfs_net_salary/52
mfs_net_hourly = mfs_net_salary/52/40
mfs_net_salary = mfs_net_salary

##Head of Household federal tax marginal rate     

hh_net_salary = 0

if((filing_status == 'HH' or filing_status == 'hh') and self_income <= s_max1):
    s_net_salary = ((self_income)*(1-fica))-(s_1*(self_income))
elif((filing_status == 'HH' or filing_status == 'hh') and self_income <= s_max2):
    s_net_salary = ((self_income)*(1-fica))-(s_1*s_max1)-(((self_income)-s_max1)*s_2)
elif((filing_status == 'HH' or filing_status == 'hh') and self_income <= s_max3):
    s_net_salary = ((self_income)*(1-fica))-(s_1*s_max1)-(s_2*(s_max2-s_max1))-(((self_income)-s_max2)*s_3)
elif((filing_status == 'HH' or filing_status == 'hh') and self_income <= s_max4):
    s_net_salary = ((self_income)*(1-fica))-(s_1*s_max1)-(s_2*(s_max2-s_max1))-(s_3*(s_max3-s_max2))-(((self_income)-s_max3)*s_4)


hh_net_monthly = hh_net_salary/12
hh_net_weekly = hh_net_salary/52
hh_net_hourly = hh_net_salary/52/40
hh_net_salary = hh_net_salary
    
#if spouse:    
#    mfj_net_salary = round(mfj_net_salary - self_401k_cont*pay_freq - spouse_401k_cont*pay_freq,2)
#else:
#    mfj_net_salary = round(mfj_net_salary - self_401k_cont*pay_freq,2)

net_salary = s_net_salary + mfj_net_salary + mfs_net_salary + hh_net_salary
net_salary_monthly = round(net_salary/12, 2)
net_salary_daily = round(net_salary/365.25, 2)

```
